```python
import qutip_qip
```

```python
import qutip_qip.circuit


qc = qutip_qip.circuit.QubitCircuit(2)
```

```python
qc.add_gate("RX", targets=[0], arg_value=0.2)
qc.add_gate("RY", targets=[0], arg_value=0.1)
qc.add_gate("RY", targets=[0], arg_value=0.1)

qc.add_gate("RZ", targets=[0], arg_value=0.3)
qc.add_gate("CNOT", targets=[0], controls=[1])
```

```python
qutip_qip.circuit.QubitCircuit.draw(qc)
```

```python
from qutip_qip.circuit.optimise import optimise


new_qc = optimise(qc)
```

```python
qutip_qip.circuit.QubitCircuit.draw(new_qc)
```
